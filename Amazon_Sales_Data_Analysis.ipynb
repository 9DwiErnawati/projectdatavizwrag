### H. Handling nilai Rupe ke Rupiah

Diskon dan harga aktual dalam Rupee. Konversikan Rupee India ke dalam dolar AS. Nilai tukar Rupee ke dolar adalah 0,012.

```python
# Nilai tukar dari Rupee India ke Rupiah Indonesia
# Nilai ini mungkin berubah-ubah, jadi pastikan untuk memeriksa nilai tukar terkini
exchange_rate = 0.012

# Konversi harga dari Rupee India ke Rupiah Indonesia
amazon['discounted_price'] = amazon['discounted_price'].apply(lambda x: x * exchange_rate)
amazon['actual_price'] = amazon['actual_price'].apply(lambda x: x * exchange_rate)

# Bulatkan harga ke 2 angka desimal
amazon['discounted_price'] = round(amazon['discounted_price'], 2)
amazon['actual_price'] = round(amazon['actual_price'], 2)
```

# 9.Membuat kolom yang mengurutkan produk berdasarkan skor penilaian.
Kolom peringkat akan berisi peringkat berdasarkan peringkat produk. Setiap peringkat akan mewakili rentang peringkat (1-2 bintang = Buruk).

```python
# Buat perulangan for dan pernyataan if else untuk mengkategorikan skala peringkat.
# Gunakan perintah .append() untuk menambahkan kategori peringkat ke daftar peringkat.

ranking = []

for score in amazon['rating']:
    if score <= .9 : ranking.append('Very Poor')
    elif score <= 1.9 : ranking.append('Poor')
    elif score <= 2.9 : ranking.append('Average')
    elif score <= 3.9 : ranking.append('Good')
    elif score <= 4.9 : ranking.append('Very Good')
    elif score == 5.0 : ranking.append('Excellent')

# Tambahkan daftar peringkat sebagai kolom peringkat dalam bingkai data amazon.
# Gunakan perintah .astype() untuk mengonversi nilai dalam kolom peringkat menjadi tipe data kategori.

amazon['ranking'] = ranking

amazon['ranking'] = amazon['ranking'].astype('category')
```
